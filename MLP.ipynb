```python
import pandas as pd
import os
from google.colab import drive #เชื่อม Google drive
drive.mount('/content/drive')
path = '/content/drive/My Drive/project' #เข้า drive
data_X = pd.read_csv(os.path.join(path,'first_data.csv'))
data_Y = pd.read_csv(os.path.join(path,'dataset2.csv'))
```

```python
data_X.drop('Unnamed: 0',1,inplace = True)
data_Y.dropna(subset=['telomere length (kb) '], inplace = True)
data_Y.drop(['No.', 'LabNo','DHEAs (BRIA)'], axis=1,inplace = True)
data_Y.drop(data_Y.columns[2: 10], axis=1,inplace = True)
data_Y.drop(data_Y.columns[27:37], axis=1,inplace = True)
```

## First Data

```python
data1 = data_Y['telomere length (kb) ']
dtset1 = data_X.copy()
dtset1['telomere length (kb) '] = list(data1)
dtset1
```

```python
dtset1.isna().sum()
```

```python
y = dtset1['telomere length (kb) '].values
X = dtset1.drop('telomere length (kb) ',1).values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=216)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
```

```python
model = MLPRegressor(
                      activation = 'relu',
                       solver = 'adam')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
import numpy as np
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R^2',metrics.r2_score(y_test, y_pred))
```

## New Section

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'hidden_layer_sizes': [10,20,50,100],
    'max_iter': [50, 100,200,500,1000],'batch_size':[4,8,16,32],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant','adaptive'],'learning_rate_init':[0.001,0.01,0.1],
    'random_state':[216]
}
```

```python
grid = GridSearchCV(model, param_grid)
grid.fit(X_train, y_train)

print(grid.best_params_) 
```

```python
grid_predictions = grid.predict(X_test) 
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, grid_predictions))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, grid_predictions))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, grid_predictions)))
print('R^2',metrics.r2_score(y_test, grid_predictions))
```
