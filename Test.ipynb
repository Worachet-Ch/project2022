```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    r2_score, 
    mean_squared_error, 
    mean_absolute_error, 
    mean_absolute_percentage_error)
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/TAUTOLOGY-EDUCATION/Deep-Learning-the-Series-2nd/main/DL101/Main%20Content/Code%20-%20Advanced%20Workshop/01.%20CAR%20PRICE/Elastic%20Net/car_price_dataset.csv')
```

```python
dataset
```

```python
dataset.drop(['name','seller_type'],1,inplace = True)
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
dataset['fuel'].replace({'Petrol':1,'Diesel':2},inplace = True)
dataset['transmission'].replace({'Manual':1,'Automatic':2},inplace = True)
dataset['owner'].replace({'First Owner':1,'Second Owner':2,'Third Owner':3,'Fourth & Above Owner':4
                         ,'Test Drive Car':5},inplace = True)
```

```python
dataset
```

```python
dataset['fuel'] = dataset['fuel'].map({1: 'Petrol', 2: 'Diesel'})
dataset['transmission'] = dataset['transmission'].map({1: 'Manual', 2: 'Automatic'})
dataset['owner'] = dataset['owner'].map({1: 'First Owner', 2: 'Second Owner', 3: 'Third Owner',4:'Fourth & Above Owner',5:'Test Drive Car'})
dataset
```

```python
dataset = pd.get_dummies(dataset, columns=['fuel'], prefix='', prefix_sep='')
dataset = pd.get_dummies(dataset, columns=['transmission'], prefix='', prefix_sep='')
dataset = pd.get_dummies(dataset, columns=['owner'], prefix='', prefix_sep='')
dataset.tail()
```

```python
X = dataset.drop(columns=['selling_price'])
y = dataset[['selling_price']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=100)
```

```python
model = Sequential()

model.add(Dense(11, activation='relu', input_shape=(11,)))
model.add(Dense(11, activation='relu'))
model.add(keras.layers.Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train, epochs=100,validation_data=(X_test, y_test))
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```
